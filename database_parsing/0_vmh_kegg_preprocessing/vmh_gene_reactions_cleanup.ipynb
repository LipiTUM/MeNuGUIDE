```python
import pandas as pd
import os
```

```python
vmh_folder = '/path/to/vmh/data/folder/'

reactions_file_path = os.path.join(vmh_folder, "vmh_reactions.csv")
gene_reactions_file_path = os.path.join(vmh_folder, "vmh_genereactions_all.csv")
```

```python
reactions = pd.read_csv(reactions_file_path)
gene_reactions = pd.read_csv(gene_reactions_file_path)
```

```python
gene_reactions
```

```python
gene_reactions_small = gene_reactions[['abbreviation', 'searched_gene_number']]
```

```python
gene_reactions_small
```

```python
df_merge = pd.merge(reactions, gene_reactions_small, how="inner", on="abbreviation")
print("Number of matched microbe reactions: " + str(df_merge.abbreviation.count()))
```

```python
gene_reactions_small.rename(columns={'abbreviation': 'reaction_abbreviation', 'searched_gene_number': 'gene_number'}, inplace=True)
```

```python
gene_reactions_small
```

```python
reaction_genes_file_path = os.path.join(vmh_folder, "vmh_reactiongenes_all.csv")
reaction_genes = pd.read_csv(reaction_genes_file_path)
```

```python
reaction_genes
```

```python
merged_left = pd.merge(reactions, gene_reactions_small, how="left", left_on="abbreviation", right_on="reaction_abbreviation")
unmatched = merged_left[merged_left.gene_number.isnull()]
print("Number of unmatched reactions in all_reactions: " + str(unmatched.abbreviation.count()))
```

```python
df_merge2 = pd.merge(reactions, reaction_genes, how="inner", left_on="abbreviation", right_on="reaction")
print("Number of matched reactions: " + str(df_merge2.abbreviation_x.count()))
```

```python
df_merge2_small = df_merge2[['abbreviation_x', 'gene_number']]
```

```python
df_merge2_small
```

```python
df_merge2_small.rename(columns={'abbreviation_x': 'reaction_abbreviation'}, inplace=True)
```

```python
df_merge2_small
```

```python
gene_reactions_small = gene_reactions_small.append(df_merge2_small, ignore_index=True)
```

```python
gene_reactions_small
```

```python
matched_reactions = pd.DataFrame(gene_reactions_small.reaction_abbreviation.unique())
matched_reactions.columns = ['reaction_abbreviation']
```

```python
matched_reactions
```

```python
merged_final = pd.merge(reactions, matched_reactions, how="left", left_on="abbreviation", right_on="reaction_abbreviation")
unmatched_final = merged_final[merged_final.reaction_abbreviation.isnull()]
print("Number of unmatched reactions in all_reactions: " + str(unmatched_final.abbreviation.count()))
```

```python
len(gene_reactions_small)
```

```python
gene_reactions_small = gene_reactions_small.drop_duplicates(subset=['reaction_abbreviation', 'gene_number'], keep='last')
```

```python
len(gene_reactions_small)
```

```python
microbe_gene_reaction_file_path = os.path.join(vmh_folder, "vmh_reaction_to_microbegenes.csv")
microbe_gene_reactions = pd.read_csv(microbe_gene_reaction_file_path)
```

```python
df_merge3 = pd.merge(reactions, microbe_gene_reactions, how="inner", left_on="abbreviation", right_on='reaction_abbreviation')
print("Number of matched reactions: " + str(df_merge3.abbreviation.count()))
```

```python
df_merge3 = df_merge3[['abbreviation', 'gene_number']]
```

```python
len(df_merge3)
```

```python
df_merge3 = df_merge3.drop_duplicates(subset=['abbreviation', 'gene_number'], keep='last')
```

```python
len(df_merge3)
```

```python
df_merge3.rename(columns={'abbreviation': 'reaction_abbreviation'}, inplace=True)
```

```python
gene_reactions_small = gene_reactions_small.append(df_merge3, ignore_index=True)
```

```python
gene_reactions_small[gene_reactions_small[['reaction_abbreviation', 'gene_number']].duplicated() == True]
```

```python
matched_reactions = pd.DataFrame(gene_reactions_small.reaction_abbreviation.unique())
matched_reactions.columns = ['reaction_abbreviation']
```

```python
merged_final2 = pd.merge(reactions, matched_reactions, how="left", left_on="abbreviation", right_on="reaction_abbreviation")
unmatched_final2 = merged_final2[merged_final2.reaction_abbreviation.isnull()]
print("Number of unmatched reactions in all_reactions: " + str(unmatched_final2.abbreviation.count()))
```
